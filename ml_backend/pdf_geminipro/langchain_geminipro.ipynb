```python
from fpdf import FPDF

# Create a PDF object
pdf = FPDF()

# Add a page
pdf.add_page()

# Set font
pdf.set_font("Arial", size = 12)

# Iterate over each dictionary in the list
for i, doc in enumerate(data):
    # Write each key-value pair on a new line
    for key, value in doc.items():
        pdf.cell(200, 10, txt = f"{key}: {value}", ln = True, align = 'L')
    # Add a line break between dictionaries
    if i < len(data) - 1:
        pdf.cell(200, 10, txt = "", ln = True, align = 'L')

# Save the pdf with name .pdf
pdf.output("/content/drive/MyDrive/OCR/output.pdf")
```

### RAG Pipeline: Embedding + LLM

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=0)
context = "\n\n".join(str(p.page_content) for p in pages)
texts = text_splitter.split_text(context)
```

```python
# texts
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
```

```python
!pip install chromadb
```

```python
vector_index = Chroma.from_texts(texts, embeddings).as_retriever()
```

```python
question = "Describe the company"
docs = vector_index.get_relevant_documents(question)
```

```python
docs
```

```python
stuff_answer = stuff_chain(
    {"input_documents": docs, "question": question,"input":docs}, return_only_outputs=True
)
```

```python
pprint(stuff_answer)
```
